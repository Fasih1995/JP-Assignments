# Q4-- Write a Python program which accepts the radius of a circle from the user and compute the area.

# Q7-- Write a program which takes 5 inputs from user for different subjectâ€™s marks, total it and generate mark sheet using grades ?

```python
Eng = input("Enter Eng Marks: ")
Isl = input("Enter Isl Marks: ")
Maths = input("Enter Maths Marks: ")
Phy = input("Enter Phy Marks: ")
Chem = input("Enter Chem Marks: ")
total_marks = 500
if int(Eng) > 0 and int(Isl) > 0 and int(Maths) > 0 and int(Chem) > 0 and int(Phy) > 0:
    marks_obtain = int(Eng) + int(Isl) + int(Maths)+ int(Phy) + int(Chem)
    percentage = (marks_obtain/total_marks)*100
    if percentage >=81.0 and percentage <=100:
        print(f"Your Marks are {marks_obtain} Your Grade is A+")
    elif percentage >=71 and percentage <81:
        print(f"Your Marks are {marks_obtain} Your Grade is A")
    elif percentage >=61 and percentage <71:
        print(f"Your Marks are {marks_obtain} Your Grade is B")
    elif percentage >=51 and percentage <61:
        print(f"Your Marks are {marks_obtain} Your Grade is C")
    elif percentage >=41 and percentage <51:
        print(f"Your Marks are {marks_obtain} Your Grade is D")
    elif percentage >=33 and percentage <41:
        print(f"Your Marks are {marks_obtain} Your Grade is E")
    elif percentage >=0 and percentage <33:
        print("Fail")
    else:
        print("Invalid Input!")
else:
    print("Input a number greter than 0")
```

# Q8-- Write a program which take input from user and identify that the given number is even or odd?
